# Heroes of PyMoli Data Analysis (Homework)

```python
import os
import pandas as pd
```

```python
purchase_data = os.path.join("generated_data","purchase_data_3.csv")
purchase_data_pd = pd.read_csv(purchase_data)
purchase_data_pd.head()
```

```python
#Player Count
player_data = os.path.join("generated_data","players_complete.csv")
player_data_pd = pd.read_csv(player_data)
player_count = player_data_pd["Player ID"].count()

#Total Number of Players
player_count_df = pd.DataFrame({"Total Players": [player_count]})
player_count_df
```

```python
#Purchasing Analysis (Total)
item_data = os.path.join("generated_data", "items_complete.csv")
item_data_pd = pd.read_csv(item_data)

#Number of Unique Items
unique_items = len(item_data_pd["Item Name"].unique())
unique_items

#Average Purchase Price
average_pp = item_data_pd["Price"].mean()
average_pp
#Total Number of Purchases
number_purchase = purchase_data_pd["Price"].count()
number_purchase
#Total Revenue
total_revenue = purchase_data_pd["Price"].sum()
total_revenue

summary = pd.DataFrame({
    "Number of Unique Items": unique_items,
    "Average Purchase Price": average_pp,
    "Total Number of Purchases": number_purchase,
    "Total Revenue": [total_revenue]})
summary
```

```python
#Gender Demographics
gender_count = player_data_pd["Gender"].value_counts()
total_count = player_data_pd["Gender"].count()

#Percentage and Count of Male Players
#Percentage and Count of Female Players
#Percentage and Count of Other / Non-Disclosed

gender_count_df = pd.DataFrame({"Total Count": gender_count})
gender_count_df["Percentage of Total"] = (gender_count/total_count*100).map("{:.2f}%".format)
gender_count_df

```

```python
#Purchasing Analysis (Gender)

#The below each broken by gender

#Purchase Count
purchase_count = purchase_data_pd["Gender"].value_counts()
#Average Purchase Price
grouped = purchase_data_pd.groupby(["Gender"])
grouped_mean = grouped["Price"].mean()
#Total Purchase Value
group_total = grouped["Price"].sum()
#Normalized Totals

summary_group = pd.DataFrame({
    "Purchase Count": purchase_count,
    "Average Purchase Pice": grouped_mean.map("${:.2f}".format),
    "Total Purchase Value": group_total.map("${:.2f}".format),
    "Normalized Totals": (group_total/total_count).map("${:.2f}".format)
})

summary_group
```

```python
#Age Demographics
age_max = player_data_pd["Age"].max()
age_min = player_data_pd["Age"].min()

age_df = pd.DataFrame({
    "Highest Age": [age_max],
    "Youngest Age": [age_min]
})
print(age_df)
age_df = pd.DataFrame(player_data_pd)

bins = [0, 9, 14, 19, 24, 29, 34, 39, 44, 50]
ages = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39","40-44", "45+"]
#The below each broken into bins of 4 years (i.e. <10, 10-14, 15-19, etc.)

age_df["Age Group"] = pd.cut(age_df["Age"], bins, labels=ages)

age_df = age_df.groupby("Age Group")
age_count = age_df["Player ID"].count()
age_total_count = player_data_pd["Player ID"].count()

summary_age_df = pd.DataFrame({"Total Count": age_count,
                              "Percentage of Total": (age_count/age_total_count*100).map("{:.2f}%".format)})
summary_age_df

```

```python
purchase_age_max = purchase_data_pd["Age"].max()
purchase_age_min = purchase_data_pd["Age"].min()
purchase_age_df = pd.DataFrame({
    "Highest Age": [purchase_age_max],
    "Youngest Age": [purchase_age_min]
})
print(purchase_age_df)
purchase_age_df = pd.DataFrame(purchase_data_pd)

bins = [0, 9, 14, 19, 24, 29, 34, 39, 44]
purchase_ages = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39","40-44"]

purchase_age_df["Age Group"] = pd.cut(purchase_age_df["Age"], bins, labels=purchase_ages)

purchase_age_df = purchase_age_df.groupby("Age Group")


#Purchase Count
purchase_age_count = purchase_age_df["Price"].count()
#Average Purchase Price
purchase_age_average_pp = purchase_age_df["Price"].mean()
#Total Purchase Value
purchase_age_total = purchase_age_df["Price"].sum()


#Normalized Totals

summary_age = pd.DataFrame({
    "Purchase Count": purchase_age_count,
    "Average Purchase Price": purchase_age_average_pp.map("${:.2f}".format),
    "Total Purchase Value": purchase_age_total.map("${:.2f}".format),
    "Normalized Totals": (purchase_age_total/total_count).map("${:.2f}".format)
})

summary_age = summary_age[["Purchase Count", "Average Purchase Price", "Total Purchase Value", \
                           "Normalized Totals"]]
summary_age
```

```python
#Top Spenders

#Identify the the top 5 spenders in the game by total purchase value, then list (in a table):


spenders_df = pd.DataFrame(purchase_data_pd)
grouped_spenders_df = spenders_df.groupby("SN")

grouped_spenders_total = grouped_spenders_df["Price"].sum()
grouped_spenders_count = grouped_spenders_df["Price"].count()
grouped_spenders_average = grouped_spenders_df["Price"].mean()


grouped_spenders_average

spenders_summary = pd.DataFrame({
    "Purchase Count": grouped_spenders_count,
    "Average Purchase Price": grouped_spenders_average.map("${:.2f}".format),
    "Total Purchase Value": grouped_spenders_total.map("${:.2f}".format)
})

top_spenders = spenders_summary.sort_values("Total Purchase Value", ascending=False)
top_spenders.iloc[0:5]

#SN
#Purchase Count
#Average Purchase Price
#Total Purchase Value
```

```python
#Most Popular Items

#Identify the 5 most popular items by purchase count, then list (in a table):
items_df = pd.DataFrame(purchase_data_pd)
grouped_items_df = items_df.groupby(["Item ID", "Item Name"])

grouped_items_total = grouped_items_df["Price"].sum()
grouped_items_count = grouped_items_df["Price"].count()
grouped_items_price = grouped_items_df["Price"].sum()/grouped_items_df["Price"].count()



items_summary = pd.DataFrame({

    "Purchase Count": grouped_items_count,
    "Total Purchase Value": grouped_items_total.map("${:.2f}".format),
    "Item Price": grouped_items_price.map("${:.2f}".format)
})

top_items = items_summary.sort_values("Purchase Count", ascending=False)
top_items.iloc[0:5]

#Item ID
#Item Name
#Purchase Count
#Item Price
#Total Purchase Value
```

```python
#Most Profitable Items

#Identify the 5 most profitable items by total purchase value, then list (in a table):
items_df = pd.DataFrame(purchase_data_pd)
grouped_items_df = items_df.groupby(["Item ID", "Item Name"])

grouped_items_total = grouped_items_df["Price"].sum()
grouped_items_count = grouped_items_df["Price"].count()
grouped_items_price = grouped_items_df["Price"].sum()/grouped_items_df["Price"].count()



items_summary = pd.DataFrame({
    "Purchase Count": grouped_items_count,
    "Total Purchase Value": grouped_items_total.map("${:.2f}".format),
    "Item Price": grouped_items_price.map("${:.2f}".format)
})

top_items = items_summary.sort_values("Total Purchase Value", ascending=False)
top_items.iloc[0:5]
#Item ID
#Item Name
#Purchase Count
#Item Price
#Total Purchase Value
```
