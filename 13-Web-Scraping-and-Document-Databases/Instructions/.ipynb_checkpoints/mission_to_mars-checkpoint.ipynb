```python
# NASA Mars News

# Dependencies
from bs4 import BeautifulSoup as bs
import requests
```

```python
url = 'https://mars.nasa.gov/news/?page=0&per_page=40&order=publish_date+desc%2Ccreated_at+desc&search=&category=19%2C165%2C184%2C204&blank_scope=Latest'
```

```python
response = requests.get(url)
print(response)
```

```python
soup = bs(response.text, 'html.parser')
print(soup.prettify())
```

```python
results = soup.find_all('div', class_='content_title')
#print(results)

news_title = []

for title in results:
    if(title.a):
        if(title.a.text):
            news_title.append(title.a.text.strip())
print(news_title)
```

```python
for x in range(len(news_title)):
    print(news_title[x])
```

# JPL Mars Space Images - Featured Image

```python
from splinter import Browser
from bs4 import BeautifulSoup
```

```python
executable_path = {'executable_path': '/usr/local/bin/chromedriver'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://www.jpl.nasa.gov/spaceimages/?search=&category=Mars'
browser.visit(url)
```

```python
browser.click_link_by_partial_text('FULL IMAGE')
```

```python
browser
```

```python
browser.click_link_by_partial_text('more info')
```

```python
html = browser.html
soup = BeautifulSoup(html, 'html.parser')

result = soup.find('figure', class_='lede')

featured_image_url = "https://www.jpl.nasa.gov" + str(result.find('a')['href'])

print(featured_image_url)
```

# Mars Hemispheres

```python
import time
```

```python
url = "https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars"
```

```python
response = requests.get(url)
print(response)
```

```python
soup = bs(response.text, 'html.parser')
print(soup.prettify())
```

```python
hem_results = soup.find_all('a', class_='itemLink product-item')
hem_results
```

```python
links = []

for title in hem_results:
    if(title.h3):
        if(title.h3.text):
            links.append(title.h3.text)
print(links)
```

```python
hemisphere_image_urls = []
```

```python
for x in range(len(links)):

    executable_path = {'executable_path': '/usr/local/bin/chromedriver'}
    browser = Browser('chrome', **executable_path, headless=False)

    url = 'https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars'
    browser.visit(url)

    browser.click_link_by_partial_text(links[x])
    title = links[x][0:-9]
    
    images_html = browser.html
    images_soup = bs(images_html, "html.parser")
    img_url = images_soup.find('div', class_='downloads').find('li').a['href']
    
    #img_url = "https://astrogeology.usgs.gov" + str(soup.find('div', class_='wide-image-wrapper'))
    image_dict = {}
    image_dict['title'] = title
    image_dict['img_url'] = img_url
    
    hemisphere_image_urls.append(image_dict)
```
